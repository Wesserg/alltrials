```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


```

```python
# Load an example pandas dataset
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
# make a summary plot of the data
fig, ax = plt.subplots()
df.plot(ax=ax)
ax.set_title('Iris Data Set')
plt.show()
```

```python
df
```

```python
# Perform a PCA analysis of the data and plot the results
from sklearn.decomposition import PCA
df_num = df.drop('species', axis=1)
pca = PCA(n_components=2)
pca.fit(df_num)
df_pca = pca.transform(df_num)
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
df_pca.plot()
plt.show()
```
